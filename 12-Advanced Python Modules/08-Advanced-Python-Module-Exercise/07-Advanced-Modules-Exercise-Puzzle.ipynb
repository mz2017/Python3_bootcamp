```python
import zipfile
```

```python
zf = zipfile.ZipFile('unzip_me_for_instructions.zip','r')
```

```python
zf.extractall('extracted_content')
```

```python
import os
```

```python
os.getcwd()
```

```python
import re

def hasPhone(text):
    phone = re.search(r'\d{3}-\d{3}-\d{4}',text)
    if phone is None:
        # print('No phone number found')
        return False
    else:
        print('Found phone number: {}'.format(phone.group()))
        return True

S = 'Here is my numbetr: 339-222-3333'
hasPhone(S)
```

```python
for folder , sub_folders , files in os.walk("extracted_content"):
    
    print("Currently looking at folder: "+ folder)
    print('\n')
    print("THE SUBFOLDERS ARE: ")
    for sub_fold in sub_folders:
        print("\t Subfolder: "+sub_fold )
    
    print('\n')
    
    print("THE FILES ARE: ")
    for f in files:
        fullpath = folder + "\\" + f;
        with open(fullpath) as file:
            content = file.readline()
            if hasPhone(content):
                print("\t File {} has phone number in it.".format(fullpath))
            else:
                print("\t No phone number found in File {}".format(fullpath))
    print('\n')
```
